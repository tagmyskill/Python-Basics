# Python Lists
The list is a most versatile datatype available in Python which can be written as a list of comma-separated values (items) between square brackets. Important thing about a list is that items in a list need not be of the same type.

```python
[1,2,3]
```

```python
['hi',1,[1,2]]
```

# Accessing Values in Lists
To access values in lists, use the square brackets for slicing along with the index or indices to obtain value available at that index.

```python
my_list = ['a','b','c']
```

```python
my_list[0]
```

```python
my_list[1:]
```

```python
my_list[:2]
```

# List Methods 

```python
# Append list 
aList = [123, 'xyz', 'zara', 'abc'];
aList.append( 2009 );
print("Updated List : ", aList)
```

```python
# list index 
aList = [321, 'xyz', 'zara', 'abc'];
print("Index for xyz : ", aList.index( 'xyz' ))
print("Index for zara : ", aList.index( 'zara' )) 
```

```python
# list reverse
aList = [123, 'xyz', 'zara', 'abc', 'xyz'];
aList.reverse();
print("List : ", aList)
```

```python
# length list 
list_len = [1,2,3,4,5,9,7]
len(list_len) 
max(list_len) # get the max from the list 
min(list_len) # get the min from the list 
```

# Delete List Elements
To remove a list element, you can use either the del statement if you know exactly which element(s) you are deleting or the remove() method if you do not know.

```python
#looping through the lists 

thislist = ["product1", "product2", "product3"]
thislist.remove("product3")
print(thislist)
```

```python
#The pop() method removes the specified index,
productlist = ["product1", "product2", "product3"]
productlist.pop()
print(productlist)
```

```python
# Empty the list 
productlist = ["product1", "product2", "product3"]
productlist.clear()
print(productlist)
```

# copying and merging the lists

```python
# copy a list 
productlist = ["product1", "product2", "product3"]
mylist = productlist.copy()
print(mylist)
```

```python
# joining two lists
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
```

```python
# loop through the list 
itemlist = ["item1", "item2", "item3"]
for x in itemlist:
  print(x)
```
