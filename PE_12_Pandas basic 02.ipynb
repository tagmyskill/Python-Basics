```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html
# identify min and max values in pandas 
df['Year'].max()  # Get the max value form the data "year"
df['Emission'].min()  # Lowest emissions 
df['Emission'].std()
```
