## Lamda expressions

In Python, lambda expressions (or lambda forms) are utilized to construct anonymous functions.
To do so, you will use the lambda keyword (just as you use def to define normal functions).

Why lambda functions?
As you will see in the next section, lambdas are treated the same as regular functions at the interpreter level. In a way, you could say that lambdas provide compact syntax for writing functions which return a single expression.

As a best practice, you need to remember that simple code is always better than complex code

Lambdas vs. Regular functions

The primary difference between a lambda and a regular function is that the lambda function evaluates only a single expression and yields a function object.

```python
def times2(var):
    return var*2
```

```python
times2(2)
```

we have to define a name for the function which returns the result when we call it. A lambda function doesn't contain a return statement because it will have only a single expression which is always returned by default. You don't even have to assign a lambda either as it can be immediately invoked

lambda p1, p2: expression 
Here, p1 and p2 are the parameters which are passed to the lambda function. You can add as many or few parameters as you need.

```python
adder = lambda x, y: x * y
print (adder (3, 2))
```

```python
# lambda's in filters 
sequences = [10,2,8,7,5,4,3,11,0, 1]
filtered_result = filter (lambda x: x % 2 == 0 , sequences) 
print(list(filtered_result))
```

lambdas in map()
the map function is used to apply a particular operation to every element in a sequence. Like filter(), it also takes 2 parameters:

```python
sequences = [10,2,8,7,5,4,3,11,0, 1]
filtered_result = map (lambda x: x*2, sequences) 
print(list(filtered_result))
```

## Map filters 

Map, Filter, and Reduce are paradigms of functional programming. They allow the programmer (you) to write simpler, shorter code, without neccessarily needing to bother about intricacies like loops and branching.

The map() function in python has the following syntax:

map(func, *iterables)

Where func is the function on which each element in iterables (as many as they are) would be applied on. Notice the asterisk(*) on iterables? It means there can be as many iterables as possible, in so far func has that exact number as required input arguments.

```python
def times2(var):
    return var*2
```

```python
seq = [1,2,3,4,5]
```

```python
map(times2,seq)
```

```python
list(map(times2,seq))
```

```python
my_skills = ['python', 'spark', 'sql', 'matplots']

uppered_skills = list(map(str.lower, my_skills))

print(uppered_skills)
```

```python
list(map(lambda var: var*2,seq))
```
